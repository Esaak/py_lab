# № 1

```python
for i in range(1, 4):    
    file_path = './lunar_images/lunar0' + str(i) + '_raw.jpg'
    new_file_path = 'updated_data' + str(i)
    img = Image.open(file_path)
    data = np.array(img)
    #interpolating coefficient
    updated_data = data - np.min(data)
    koef = 255/np.max(updated_data)
    updated_data = updated_data * koef
    pil_image = Image.fromarray(updated_data).convert('RGB')
    pil_image.save('Pil_image' + new_file_path + '.png')
```

# № 2

```python
def average(data):
    return np.average(data)
```

```python
def draw(x, y, new_file_path):
    plt.figure(figsize=(12, 7))
    plt.minorticks_on()
    plt.grid(
        which='major'
    )
    plt.grid(
        which='minor',
        linestyle='--'
    )
    plt.plot(x, y, color='green')
    plt.savefig(new_file_path)
```

```python
for i in range(1, 4):
    file_path = './signals/signal0' + str(i) + '.dat'
    new_file_path = 'updated_plot' + str(i) + '.png'
    data = []
    with open(file_path, 'r') as f:
            numbers = f.readlines()
            for i in numbers:
                data.append(float(i))
    data = np.array(data)
    updated_data = data 
    for i in range(1, 10):
        updated_data[i] = average(updated_data[:i+1])
    for i in range(10, len(data)):
        updated_data[i] = average(updated_data[i-9:i+1]) 
    x = np.arange(0, len(updated_data))
    draw(x, updated_data, new_file_path)
```

```python


def draw_plot_for_gif(x, y, ax):
    image = ax.scatter(x, y, animated = True)
    return image

def do_gif(u0, A, N):
    fig, ax = plt.subplots()
    ax.grid(True)
    ax.minorticks_on()
    images_list = []
    
    for i in range(N):
        image = draw_plot_for_gif(np.arange(0, len(u0)), u0, ax)
        images_list.append([image])
        
        u0 = u0 - 0.5 * np.matmul(A, u0)
    #make gif from plots list
    ani = animation.ArtistAnimation(fig, images_list)
    ani.save('movie.gif')
```

```python
file_path = 'start.dat'
N = 255
data = []
with open(file_path, 'r') as f:
    numbers = f.readlines()
    for i in numbers:
        data.append(float(i))
    f.close()
data = np.array(data)
A = np.eye(len(data))

B0 = np.eye(len(data)-1) *(-1)
row_0 = np.zeros(len(data)-1)
column_0 = np.zeros(len(data)).reshape(50,1)
B0 = np.vstack((row_0, B0))
B0 = np.hstack((B0, column_0))

A+=B0
do_gif(data, A, N)
```
